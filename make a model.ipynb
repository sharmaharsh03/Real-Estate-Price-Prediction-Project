```python
# load a basic library
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
```

```python
# load a data
df=pd.read_csv("Bengaluru_House_Data_usualdata.csv")
df
```

```python
df.location = df.location.apply(lambda x: x.strip())
```

```python
df.shape
```

```python
# in location value in char. now convert in numbers 
dummies = pd.get_dummies(df.location)
print(dummies)
```

```python
dff=pd.concat([df,dummies.drop("other",axis="columns")], axis="columns" )
print(dff.head(5))
```

```python
dff = dff.drop(['location','size','price_per_sqft'],axis='columns')
```

```python
dff
```

```python
X=dff.drop("price",axis="columns")
X
```

```python
y=dff.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
